# State wise population analysis

```python
# import library
import selenium
import pandas as pd 
import warnings
from selenium import webdriver
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
```

```python
# import libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_population')
```

```python
data=[]
```

```python
datas=driver.find_elements(By.TAG_NAME,'td')
for i in datas:
    info=i.text
    data.append(info)
data
```

```python
clean_data=data[3:]
```

```python
clean_data
```

```python
clean_state=clean_data[1:1000:14]
```

```python
print(clean_state)
```

```python
population_2011=clean_data[2:1000:14]
print(population_2011)
```

```python
growth=clean_data[4:1000:14]
print(growth)
```

```python
population_2023=clean_data[5:1000:14]
print(population_2023)
```

```python
df_population=pd.DataFrame({"State":clean_state,"Population 2011":population_2011,"Population 2023":population_2023,"Population Growth":growth})
df_population
```

There are many problems with this dataset so now we fix all problems

# Cleaning the data

```python
columns_to_clean = ['Population 2011', 'Population Growth', 'Population 2023']

# Loop through each column and apply cleaning operations
for col in columns_to_clean:
    df_population[col] = df_population[col].str.rstrip('%')  # Remove trailing '%'
    df_population[col] = df_population[col].str.replace('−', '-')  # Replace Unicode minus with standard minus
    df_population[col] = df_population[col].str.replace(',', '')  # Remove commas
    df_population[col] = df_population[col].str.extract('(\d+)', expand=False)  # Extract numeric values
    df_population[col] = pd.to_numeric(df_population[col])  # Convert to numeric
```

Now at this time all problem fix

# Top 10 population growth state

```python
# Sort the DataFrame by 'Population Growth' in descending order and select the top 10
top_10_growth = df_population.sort_values(by='Population Growth', ascending=False).head(10)
top_10_growth
```

The highest population growth has occurred in Dadra and Nagar Haveli and Daman and Diu

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame (replace with your actual DataFrame)
data = {
    'State': ['Dadra and Nagar Haveli and Daman and Diu', 'Puducherry', 'Meghalaya', 'Arunachal Pradesh', 'Bihar'],
    'Population Growth': [55.1, 28.1, 27.9, 26.0, 25.4]  # Example growth rates
}
df_pop = pd.DataFrame(data)

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(df_pop['Population Growth'], labels=df_pop['State'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title("Top 5 Distribution of States with Growth Rate")
plt.ylabel('')  # Hide the y-label as it doesn't apply to pie charts

plt.show()
```

```python
# Sort the DataFrame by 'Population Growth' in descending order and select the top 10
top_10_growth = df_population.sort_values(by='Population Growth', ascending=False).head(10)

# Plot Growth Rate using Seaborn with a list of colors
plt.figure(figsize=(12,8))
sns.barplot(x='State', y='Population Growth', data=top_10_growth, palette='Blues_d')
plt.xlabel('State')
plt.ylabel('Population Growth (%)')
plt.title('Top 10 States by Population Growth Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Cleaning the data

```python
columns_to_clean = ['Area (in km sqr)','Percent']

# Loop through each column and apply cleaning operations
for col in columns_to_clean:
    df_area[col] = df_area[col].str.rstrip('%')  # Remove trailing '%'
    df_area[col] = df_area[col].str.replace('−', '-')  # Replace Unicode minus with standard minus
    df_area[col] = df_area[col].str.replace(',', '')  # Remove commas
    df_area[col] = df_area[col].str.extract('(\d+)', expand=False)  # Extract numeric values
    df_area[col] = pd.to_numeric(df_area[col])  # Convert to numeric
```
